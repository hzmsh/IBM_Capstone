```python
import pandas as pd
import numpy as np



from geopy.geocoders import Nominatim 

import requests 
from pandas.io.json import json_normalize 


import matplotlib.cm as cm
import matplotlib.colors as colors


from sklearn.cluster import KMeans


import folium 

print('Libraries imported.')
```

```python
canadam = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

```python
df = canadam[0]
df.replace(to_replace='Not assigned', value=np.nan,inplace=True)
df.dropna(subset=['Borough'], inplace = True)
df['Neighbourhood'] = df['Neighbourhood'].fillna(df['Borough'])
df.head()
```

```python
df['Postcode'] = df['Postcode'].astype(str)
df['Borough'] = df['Borough'].astype(str)
df['Neighbourhood'] = df['Neighbourhood'].astype(str)
df = df.groupby('Postcode').agg({'Postcode':'first','Borough':'first','Neighbourhood': ', '.join})
df = df[['Postcode','Borough','Neighbourhood']].reset_index(drop=True)
df.columns = ['PostalCode','Borough','Neighbourhood']
df.head()
```

# 1st Answer

```python
df.shape
df
geo_file = "https://cocl.us/Geospatial_data"
geo_data = pd.read_csv(geo_file)
geo_data.head()
geo_data.columns = ['PostalCode', 'Latitude', 'Longitude']
geo_data
df_geo = pd.merge(df, geo_data, on='PostalCode')
df_geo.head()
```

```python
torronto_data = df_geo[df_geo['Borough'].str.contains('Toronto')].reset_index(drop=True)
```

```python
latitude = 43.7337
longitude = -79.5175


map_torronto = folium.Map(location=[latitude, longitude], zoom_start=11)


for lat, lng, label in zip(torronto_data['Latitude'], torronto_data['Longitude'], torronto_data['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_torronto)  
    
map_torronto
```
